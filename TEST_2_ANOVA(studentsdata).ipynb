```python
df = pd.read_excel(r"C:\Users\sxoma\Downloads\studentdata.xls",sheet_name='anova')
df
```

```python
control = df['Control'].dropna()
treatment1 = df['Treatment1'].dropna()
treatment2 = df['Treatment2'].dropna()
```

```python
#normality control
import scipy.stats as st
st.shapiro(control), st.shapiro(treatment1),st.shapiro(treatment2)
```

```python
#variance control
st.levene(control, treatment1,treatment2)
```

```python
#one-way anova
F , p = st.f_oneway(control,treatment1,treatment2)
F , p
```

```python
res = st.tukey_hsd(control, treatment1, treatment2)
print(res)
```

#Υπάρχει στατιστικά σημαντική διαφορά μεταξύ της ομάδας 1 και της ομάδας 2, καθώς η p-value είναι 0.040, μικρότερη από 0.05. Το διάστημα εμπιστοσύνης δεν περιλαμβάνει το 0, ενισχύοντας την ύπαρξη σημαντικής διαφοράς.
